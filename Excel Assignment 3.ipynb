1.Write a formula to check if the marks in cell  are greater than or equal to 40. If yes, return B2 "Fail" .

2.A company gives a bonus  if sales are above 40,000 and attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not

 If marks in B2 "A" are ≥ 90 → , ≥ 75 → , ≥ 50 → , otherwise FAIL.

4.Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?
=VLOOKUP("E103", A2:C5, 3, FALSE)
The Employee ID is present in the first column of the table.We want to find the Department, which is in another column to the right.The data is arranged vertically, making VLOOKUP the most suitable choice.
