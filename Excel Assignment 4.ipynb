1.What is the purpose of a PivotTable in Excel?
-A PivotTable is used to summarize, analyze, and reorganize large datasets quickly.

2.Create a PivotTable to show Total Sales by Region.
-To show total sales, go to Insert → PivotTable → New Worksheet, then drag Region to Rows and Sales to Values.

3.Apply Conditional Formatting to highlight sales greater than 400.
-For Conditional Formatting, go to Home, select the Total Sales column, then click Conditional Formatting → Greater Than, and enter the required value.

4.Explain the difference between Sorting and Filtering with an example from this dataset.
-Sorting rearranges the data and is used to rank data in ascending or descending order. Filtering shows only specific values and hides the rest of the data.

5.Create a PivotChart showing Monthly Sales Trend.
-Create a PivotTable, drag Date to Rows and Sales to Values. Then insert a chart, which automatically sets the X-axis and Y-axis.

6.Use a Slicer to filter data by Product and observe changes in sales.
- Go to the PivotTable, click PivotTable Analyze → Insert Slicer, select Product, then click OK.

7.Why are Dashboards important for decision-making in business analysis?
-Dashboards are used for quick insights and a summarized view of data.

8.Create a Dashboard that includes:
-Create an Excel dashboard by first converting the sales data into a table. Insert a PivotTable to show Sales by Region and add a column or bar chart for visualization. Use a Product slicer to filter the data interactively. Then create another PivotTable grouped by month and insert a line chart to display the monthly sales trend. Arrange both charts and the slicer neatly on a dashboard sheet for a clean, interactive view.

9.Group dates in the PivotTable into Months and calculate the average sales per month.
-Drag Date to Rows, right-click, choose Group → Months, and change Sales to Average in Value Field Settings.

10.Combine Apples and Oranges into one group called Fruits and compare its sales with other products.
-Create a PivotTable, then group Apples and Oranges to compare Apples and Oranges against all products.
