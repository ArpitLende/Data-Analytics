1.What is data cleaning, and why is it important in data analysis?
-Data cleaning is the process of removing errors, duplicates, missing values, and inconsistencies from data to make it accurate and reliable.
It is important because clean data leads to correct analysis, better decisions, and reliable results.
Analyzing unclean data can cause wrong conclusions, biased results, poor predictions, and loss of credibility.Common steps in data cleaning:
.Remove duplicate records
.Handle missing values
.Correct errors and inconsistencies
.Standardize formats (dates, units, text)

3.Explain the use of text functions such as TRIM , LEFT, RIGHT, MID, and CONCAT in data cleaning
Text functions are used in data cleaning to correct, extract, and combine text data.
.TRIM: Removes extra spaces from text (except single spaces between words).
 Use: Fixes inconsistent spacing in names or entries.
.LEFT: Extracts a specified number of characters from the beginning of text.
 Use: Gets codes or initials from the start of a text string.
.RIGHT: Extracts characters from the end of text.
 Use: Extracts suffixes, IDs, or last digits.
.MID: Extracts text from the middle of a string starting at a given position.
 Use: Pulls specific information like area codes or middle names.
.CONCAT: Combines text from multiple cells into one.
 Use: Joins first name and last name, or merges columns.
